## PRAKTIKUM/LATIHAN DI KELAS

```python
daftarNama = ["adam","isna","stip", "iksang", "sitti"]
daftarNama.append("Adit")
daftarNama.insert(3, "Rahmiati")
daftarNama.pop()
daftarNama.sort(reverse=True)
daftarNama.remove("stip")

i = 1

for nama in daftarNama:
    print("Nama Mahasiswa ke -", i, ":", nama)
    i += 1

print(daftarNama)
```

### Analisis:
Kode tersebut membuat daftar nama mahasiswa dan melakukan beberapa operasi pada list tersebut. Pertama, program menambahkan nama “Adit” ke akhir daftar, lalu menyisipkan “Rahmiati” di posisi ke-3. Setelah itu, elemen terakhir dihapus menggunakan pop(), dan daftar diurutkan secara terbalik dari Z ke A dengan sort(reverse=True). Kemudian, nama “stip” dihapus dari daftar menggunakan remove(). Selanjutnya, program menampilkan setiap nama mahasiswa dengan nomor urut menggunakan perulangan for, dan akhirnya mencetak isi daftar nama yang sudah diperbarui.

```python
daftarNama = ("adam","isna","stip", "iksang", "sitti")
print(daftarNama)
```

### Analisis:
Kode tersebut membuat tuple bernama daftarNama yang berisi lima elemen: "adam", "isna", "stip", "iksang", dan "sitti". Fungsi print(daftarNama) digunakan untuk menampilkan seluruh isi tuple. Karena tuple bersifat immutable (tidak dapat diubah setelah dibuat), data di dalamnya tidak bisa ditambah, dihapus, atau diubah seperti pada list.

```python
daftarNama = {"adam","isna","stip", "iksang", "sitti"}

i = 1
for nama in daftarNama:
    print("Nama Mahasiswa ke -", i , ": ", nama)
    i += 1
```

### Analisis:
Kode tersebut membuat set bernama daftarNama yang berisi lima elemen: "adam", "isna", "stip", "iksang", dan "sitti". Set di Python tidak memiliki urutan tetap, sehingga urutan elemen yang dicetak bisa berbeda setiap kali program dijalankan. Selanjutnya, program menggunakan perulangan for untuk menampilkan setiap nama mahasiswa dengan nomor urut menggunakan variabel i yang bertambah satu di setiap iterasi. Hasilnya akan menampilkan semua nama, namun urutannya acak karena sifat set yang unordered dan unik (tanpa duplikasi).

```python
person = {"Nama": "adam", "Hp": "012345678910", "Alamat": "Bone"}
for key, value in person.items():
    print(key, ":", value)
```

### Analisis:

### LATIHAN/TUGAS

```python
i = 0
while i <= 5:
    print(i)
    i += 1
```

### Analisis:
Kode tersebut menggunakan perulangan while untuk mencetak angka dari 0 hingga 5. Variabel i diawali dengan nilai 0, lalu selama kondisi i <= 5 masih bernilai True, program akan mencetak nilai i dan menambahkannya satu per satu dengan i += 1. Setelah i mencapai 6, kondisi menjadi False dan perulangan berhenti.

```python
i = 1
while i <= 13:
    if i == 7:           
        print("WOW")
    else:
        print(i)
    i += 2               
```

### Analisis:
Kode tersebut menggunakan perulangan while untuk mencetak angka dari 1 hingga 13 dengan kenaikan 2 setiap kali (i += 2). Pada setiap iterasi, program memeriksa apakah nilai i sama dengan 7. Jika iya, maka program mencetak "WOW", sedangkan jika tidak, program mencetak nilai i.

```python
i = 1
while i <= 10:
    if i <= 8:
        if i % 2 == 1:
            print("hidup")
        else:
            print("mahasiswa")
    else:
        print(i)
    i += 1
```

### Analisis:
Kode tersebut menggunakan perulangan while untuk mencetak teks berdasarkan nilai i dari 1 hingga 10. Selama i <= 8, program memeriksa apakah i ganjil atau genap. Jika ganjil (i % 2 == 1), maka mencetak "hidup", dan jika genap, mencetak "mahasiswa". Setelah i lebih dari 8, program akan mencetak langsung nilai i.

```python
a, b = 1, 2
while a <= 34:
    print(a)
    a, b = b, a + b
```

### Analisis:
Kode tersebut menghasilkan deret Fibonacci menggunakan perulangan while. Variabel a dan b masing-masing dimulai dengan nilai 1 dan 2. Selama a <= 34, program mencetak nilai a, lalu memperbarui kedua variabel dengan rumus a, b = b, a + b, sehingga setiap nilai berikutnya merupakan penjumlahan dua nilai sebelumnya.

```python
i = 1
while i <= 13:
    if i % 3 == 0:
        print("Ha")
    else:
        print(i)
    i += 1
```

### Analisis:
Kode tersebut menggunakan perulangan while untuk mencetak angka dari 1 hingga 13 dengan kondisi khusus. Jika i habis dibagi 3 (i % 3 == 0), maka program mencetak "Ha", sedangkan jika tidak, program mencetak nilai i.

```python
i = 1
while i <= 13:
    if i % 3 == 0:
        print("Ha")
    elif i % 5 == 0:
        print("Hi")
    else:
        print(i)
    i += 1
```

### Analisis:
Kode tersebut menggunakan perulangan while untuk mencetak angka dari 1 hingga 13 dengan kondisi khusus. Jika i adalah kelipatan 3 (i % 3 == 0), program mencetak "Ha". Jika i adalah kelipatan 5 (i % 5 == 0), program mencetak "Hi". Untuk angka lainnya, program mencetak nilai i itu sendiri.
